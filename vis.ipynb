```python
import os
import json
import random
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import cv2

def visualize_random_objects_with_area_ratio(folder, num_videos=10, frames_per_video=2):
    if not os.path.exists(folder):
        print(f"❌ 路径不存在: {folder}")
        return

    videos = [v for v in os.listdir(folder) if os.path.isdir(os.path.join(folder, v))]
    selected_videos = random.sample(videos, min(num_videos, len(videos)))

    for vid in selected_videos:
        video_path = os.path.join(folder, vid)
        frames = [f for f in os.listdir(video_path) if os.path.isdir(os.path.join(video_path, f))]
        if not frames:
            continue
        selected_frames = random.sample(frames, min(frames_per_video, len(frames)))
        for fid in selected_frames:
            print("="*60)
            print(f"🎬 Video: {vid} | Frame: {fid}")
            print("="*60)
            frame_dir = os.path.join(video_path, fid, "wide")
            img_path = os.path.join(frame_dir, "image.png")
            json_path = os.path.join(frame_dir, "instances.json")

            if not os.path.exists(img_path) or not os.path.exists(json_path):
                continue

            image_bgr = cv2.imread(img_path)
            image = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)
            H, W, _ = image.shape
            image_area = H * W

            with open(json_path, 'r') as f:
                objs = json.load(f)

            for i, obj in enumerate(objs):
                x1, y1, x2, y2 = map(int, obj["box_2d_rend"])
                category = obj.get("category", "object")
                obj_id = obj.get("id", "")[:8]
                box_area = (x2 - x1) * (y2 - y1)
                ratio = box_area / image_area

                fig, ax = plt.subplots(figsize=(10, 6))
                ax.imshow(image)
                ax.add_patch(patches.Rectangle((x1, y1), x2 - x1, y2 - y1,
                                               linewidth=2, edgecolor='lime', facecolor='none'))
                print( f"{category} | Area Ratio: {ratio:.3f}")
                # ax.text(x1, y1 - 10, f"{category} | Area Ratio: {ratio:.3f}",
                #         fontsize=10, color='lime', backgroundcolor='black')
                ax.set_title(f"Video: {vid} | Frame: {fid} | ID: {obj_id}")
                ax.axis("off")
                plt.show()

# 示例路径：请替换为实际存在的路径
filtered_obj_folder = "/home_sfs/zhouenshen/dataset/3D/cubifyanything/filter_obj"
visualize_random_objects_with_area_ratio(filtered_obj_folder, num_videos=10, frames_per_video=2)

```
